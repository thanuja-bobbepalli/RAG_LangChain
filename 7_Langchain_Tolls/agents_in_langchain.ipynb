```python
from langchain_core.tools import tool
import requests
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
search_tool=DuckDuckGoSearchRun()
```

```python
@tool
def get_weather_data(city: str)-> str:
    """
    This function fetches the current weather data for a given city
    """
    url = f'https://api.weatherstack.com/current?access_key=87a85acae54989ec357486912b2e1a3e&query={city}'
    
    response = requests.get(url=url)

    return response.json()
```

```python
get_weather_data.invoke({"city": "vijayawada"})
```

```python
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
```

```python
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.3-70B-Instruct",
    task="text-generation",
    temperature=0
)

llm = ChatHuggingFace(llm=llm)
```

```python
from langchain_classic.agents import create_react_agent, AgentExecutor
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.tools import tool
import langchainhub as hub
```

```python
from langsmith import Client

# Initialize the client
client = Client()

# Pull the prompt using pull_prompt
prompt = client.pull_prompt("hwchase17/react")
```

```python
agent = create_react_agent(
    llm=llm,
    tools=[search_tool, get_weather_data],
    prompt=prompt
)
```

```python
# Step 4: Wrap it with AgentExecutor
agent_executor = AgentExecutor(
    agent=agent,
    tools=[search_tool, get_weather_data],
    verbose=True
)
```

```python
response = agent_executor.invoke({'input':'find the capital of india, then find it\'s current weather condition'})
```
