## Metho2 - usnig structured Tool

```python
from langchain_community.tools import StructuredTool
from pydantic import BaseModel,Field

class MutiplyInputs(BaseModel):
    a: int=Field(required=True,description="this is the first number to add")
    b: int=Field(required=True,description="This is the second number")
    
```

```python
class MutipleInputs(BaseModel):
    a: int = Field(
    description="this is the first number to add",
    json_schema_extra={"required": True}
)
    b: int = Field(
    description="this is the second  number to add",
    json_schema_extra={"required": True}
)
```

```python
def multiply_func(a: int,b: int ) -> int:
    return a*b
```

```python
multiply_tool = StructuredTool.from_function(
    func=multiply_func,
    name="multiply",
    description="multiply two numbers",
    args_schema=MutipleInputs
)
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)
print(multiply_tool.args)
```

## Method 3 Using Base Tool class

```python

from langchain_community.tools import BaseTool
from typing import Type
```

```python
class MutipleInputs(BaseModel):
    a: int = Field(
    description="this is the first number to add",
    json_schema_extra={"required": True}
)
    b: int = Field(
    description="this is the second  number to add",
    json_schema_extra={"required": True}
)
```

```python
class MultiplyTool(BaseTool):
    name: str ="multiply"
    description:str ="multiply two numbers"
    
    args_schema:Type[BaseModel]=MutipleInputs
    
    def _run(self,a:int,b:int)->int:
        return a*b
```

```python
result = multiply_tool.invoke({'a':3, 'b':3})

print(result)
print(multiply_tool.name)
print(multiply_tool.description)

print(multiply_tool.args)
```


# Toolkit

```python

from langchain_core.tools import tool

# Custom tools
@tool
def add(a: int, b: int) -> int:
    """Add two numbers"""
    return a + b

@tool
def multiply(a: int, b: int) -> int:
    """Multiply two numbers"""
    return a * b
```

```python
class MathToolKit:
    def get_tools(self):
        return [add,multiply]
    
```

```python

toolkit = MathToolKit()

tools = toolkit.get_tools()


for tool in tools:
    print(tool.name,'=>',tool.description)
```
